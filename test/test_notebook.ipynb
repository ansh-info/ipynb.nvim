```python
# Phase 3 test - matplotlib plot (requires image.nvim)
# %matplotlib inline is auto-configured by the kernel on start.
# Do NOT call matplotlib.use() here - it overrides the inline backend.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
plt.figure(figsize=(6, 3))
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.title('ipynb image test')
plt.tight_layout()
plt.show()
```

```python
# Phase 3 test - multiple images in one cell (tests vertical stacking)
# Each plt.show() should produce a separate image stacked below the previous.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)

plt.figure(figsize=(6, 3))
plt.plot(x, np.sin(x), color='steelblue', label='sin(x)')
plt.title('Multi-image test: plot 1 of 3')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.plot(x, np.cos(x), color='tomato', label='cos(x)')
plt.title('Multi-image test: plot 2 of 3')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.plot(x, np.sin(x) * np.cos(x), color='seagreen', label='sin*cos')
plt.title('Multi-image test: plot 3 of 3')
plt.legend()
plt.tight_layout()
plt.show()
```
